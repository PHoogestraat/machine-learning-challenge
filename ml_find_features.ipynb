```python
path = "data/"
file = "exoplanet_data.csv"
path_file = path + file
df = pd.read_csv(path_file)
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```
